Seaborn

Scatter Plots

- Los gráficos de dispersión son útiles para comprender patrones, tendencias y correlaciones entre variables. Permiten identificar la presencia o ausencia de cualquier relación entre las variables. Cuando los puntos en el gráfico de dispersión forman un patrón discernible, sugiere una relación entre las variables. Diferentes patrones pueden indicar correlaciones positivas o negativas, relaciones lineales o no lineales, agrupamientos o grupos de datos, o la ausencia de cualquier relación aparente.

```python
# Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# Ignore all warning
warnings.filterwarnings('ignore')
```

```python
# Estilo de plot de Seaborn
sns.set(style='darkgrid')
```

Usaremos el data set de tips

Diccionario de Datos


*   total_bill = monto total de la cuenta
*   tip = cantidad de propina
*   sex = género del cliente
*   smoker = si el cliente fuma o no
*   day = día de la semana
*   time = monemento del dia (almuerzo o cena)
*   size = cantiad de comensales en la mesa



```python
# Carga de datos (tips)
tips = sns.load_dataset("tips")
```

```python
# Mostrar los datos
tips.head() # por defecto head () trae 5 registros
```

```python
# Uso basico del scatter plot
# Scatter Plot entre el total de pago y la propina
# Parametros
# x='total_bill': Especifica que la columna 'total_bill' debe utilizarse para el eje x.
# y='tip': Especifica que la columna 'tip' debe utilizarse para el eje y.
# data=tips: Indica el DataFrame (tips) que contiene los datos.


sns.relplot(x='total_bill', y = 'tip', data = tips)
plt.xlabel('Monto total de la Cuenta')
plt.ylabel ('Cantidad de Propina ')
plt.title('Relación entre Monto Total y Propina')
plt.show()
```

```python
# Ejemplo 2
# Scatter Plot con una nueva variable smoker
# Parametros
# x='total_bill': Esto especifica que la columna 'total_bill' se debe utilizar para el eje x.
# y='tip': Esto especifica que la columna 'tip' se debe utilizar para el eje y.
# data=tips: Esto especifica el DataFrame (tips) que contiene los datos.
# hue='smoker': Esto especifica que la columna 'smoker' se debe utilizar para diferenciar los puntos de datos por color.

sns.relplot(x = 'total_bill', y = 'tip', data=tips,hue='smoker')
plt.show()
```

```python
# Ejemplo 3
# Scatter Plot con los parametros hue y style para diferenciar puntos por las categorias
# Parametros
# x='total_bill': Especifica que la columna 'total_bill' debe utilizarse para el eje x.
# y='tip': Especifica que la columna 'tip' debe utilizarse para el eje y.
# data=tips: Indica el DataFrame (tips) que contiene los datos.
# hue='smoker': Utiliza la columna 'smoker' para diferenciar los puntos de datos por tonalidad.
# style='time': Utiliza la columna 'time' para diferenciar los puntos de datos por estilo.
sns.relplot(x='total_bill', y = 'tip', data = tips, hue= 'smoker', style ='time')
plt.show()
```
