### Line Plot

- Un gráfico de línea es un tipo de gráfico que representa puntos de datos conectados por líneas rectas. Se utiliza comúnmente para mostrar la tendencia o progresión de una variable a lo largo de un rango continuo, como el tiempo o una secuencia numérica. En un gráfico de línea, el eje x generalmente representa la variable independiente (por ejemplo, el tiempo) y el eje y representa la variable dependiente (por ejemplo, medición o valor).

- Los gráficos de línea son efectivos para visualizar el patrón general, la tendencia y la variabilidad de una variable. Nos permiten observar cómo cambian los valores a lo largo del rango de la variable independiente e identificar cualquier relación, tendencia o patrón que pueda existir. Los gráficos de línea pueden revelar información como tendencias ascendentes o descendentes, fluctuaciones, estacionalidad o la ausencia de cualquier tendencia notable.

```python
# Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
# Ejemplo 1
# Crear un DataFrame con una columa de Tiempo de tenga de 0 a 499 y una columna de valores en python
# Parametros
# np.arrange(500) = Crea una secuencia de números desde 0 a 499
# np.random.randn(500) = Genera 500 valores aleatorios provenientes de una distribución normal estandar
# media = 0 , desviación estandar = 1  utilizando la función randn de numpy.
# cumsum() =  es la suma acumulativa  de los valores aleatorios en cada posicion
# time - value = con los nombres de la  columnas en el Data Frame
# En resumen, el df contiene dos columnas "time" con los valores 0 - 499 con valores aleatorios generados
df=pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
```

```python
# Crea un gráfico de line utilizando relplot de seaborn , con el eje de x como time (tiempo) y el eje y como value (valor)
# Parámetros
# x=time: Indica que la variable en el eje x del gráfico será la columna "time" del DataFrame df
# y=value: Indica que la variable en el eje y del gráfico será la columna "value" del DataFrame df
# kind= line : Especifica que se desea crear un gráfico de líneas
# data=df : Especifica el DataFrame que contiene los datos que se utilizarán para crear el gráfico
g = sns.relplot(x="time", y="value", kind="line", data=df)
```

### Uso de datos de un conjunto de Datos
- Este conjunto de datos se utiliza comúnmente en ejemplos y tutoriales para realizar análisis y visualizaciones relacionadas con la resonancia magnética funcional (fMRI), que es una técnica de neuroimagen utilizada para medir la actividad cerebral.

```python
# Ejemplo 2
# load_dataset de Seaborn para cargar un conjunto de datos específico llamado "fmri"
fmir= sns.load_dataset("fmri")
```

```python
# Crear un gráfico de línea con el eje x como "timepoint" (punto de tiempo) y el eje y como "signal" (señal) a partir del conjunto de datos fmri.
# Parametros
# x="timepoint": Indica que la variable en el eje x del gráfico será la columna "timepoint" del DataFrame fmir. En este contexto, "timepoint" representa momentos específicos en el tiempo .
# y="signal": Indica que la variable en el eje y del gráfico será la columna "signal" del DataFrame fmir. En el contexto de fMRI, "signal" generalmente se refiere a la intensidad de la señal registrada en la resonancia magnética funcional.
sns.relplot(x="timepoint", y="signal", data=fmir, kind="line")
plt.show()
```

- Crear un gráfico de líneas con el eje x como "timepoint" (punto de tiempo) y el eje y como "signal" (señal), sin intervalo de confianza (ci=None)

```python
# Ejemplo 3
# ci=None: Configura el parámetro de intervalo de confianza (ci) en None, lo que significa que no se mostrarán intervalos de confianza alrededor de las líneas.
sns.relplot(x="timepoint", y="signal", data=fmir, kind="line", ci=None)
plt.show()
```

- Crear un gráfico de líneas con el eje x como "timepoint" (punto de tiempo) y el eje y como "signal" (señal), utilizando la desviación estándar como intervalo de confianza (ci="sd").

```python
# Ejemplo 4
# ci="sd": Configura el parámetro de intervalo de confianza (ci) para utilizar la desviación estándar ("sd").
# Esto significa que el gráfico incluirá un intervalo de confianza alrededor de las líneas que se basará en la desviación estándar de los datos. La sombra alrededor de las líneas indicará la variabilidad de los datos en relación con la desviación estándar.
sns.relplot(x="timepoint", y="signal", data=fmir, kind="line", ci="sd")
plt.show()
```
